```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt


df1 = pd.read_csv("FOOD-DATA-GROUP1.csv")
df2 = pd.read_csv("FOOD-DATA-GROUP2.csv")
df3 = pd.read_csv("FOOD-DATA-GROUP3.csv")
df4 = pd.read_csv("FOOD-DATA-GROUP4.csv")
df5 = pd.read_csv("FOOD-DATA-GROUP5.csv")

print(df1.info())
print(df2.info())
print(df3.info())
print(df4.info())
print(df5.info())
```

```python
df_combined = pd.concat([df1, df2, df3, df4, df5], ignore_index=True)
```

```python
df_combined
```

```python
# check if rows length matches
all_rows = len(df1)+len(df2)+len(df3)+len(df4)+len(df5)

print(f"Rows after concatenation: {len(df_combined)}")
if all_rows == len(df_combined):
    print("Success! No rows lost during concatenation.")
else:
    print("Mismatch detected. Investigate further.")
```

```python
if df_combined.columns.equals(df1.columns):
    print(f"Combined data columns count is {len(df_combined.columns)}\nMatchs the original data")
```

```python
duplicates = df_combined.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")

df_combined.drop_duplicates(inplace=True)
```

```python
print(df_combined.dtypes)
```

```python

df = df_combined.drop(df_combined.columns[[0, 1]], axis=1)
```

```python
df.head()
```

```python
for col in df.columns:
    missing_pct = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, missing_pct*1000))
    if missing_pct > 0.3:
        print(f"Column '{col}' has {missing_pct*100:.2f}% missing values")
```

```python
print(df.describe())
```

```python
numeric_cols = df.select_dtypes(include=['number']).columns
df[numeric_cols].boxplot(grid=False, figsize=(10, 6), color='blue')
plt.xticks(rotation=90, fontsize=8)
plt.show()
```

```python
print(f"Total unique food items: {len(df['food'].unique())}")
```

```python
df.to_csv("cleaned_data.csv", index=False)
```
